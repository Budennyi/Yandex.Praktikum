## Часть 1. Приоритизация гипотез.

В файле 9 гипотез по увеличению выручки интернет-магазина с указанными параметрами Reach, Impact, Confidence, Effort.

Файл /datasets/hypothesis.csv

- Hypothesis — краткое описание гипотезы;
- Reach — охват пользователей по 10-балльной шкале;
- Impact — влияние на пользователей по 10-балльной шкале;
- Confidence — уверенность в гипотезе по 10-балльной шкале;
- Efforts — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.


Изучим таблицу с гипотезами

```python
hypos = pd.read_csv('/datasets/hypothesis.csv')
hypos.columns  = hypos.columns.str.lower()
hypos.info()
```

Данные в порядке

#### 1. Примените фреймворк ICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета.

Формула расчета фреймворка ICE:

$$ICE = \frac{Impact * Confidence}{Efforts}$$

```python
hypos['ice'] = (hypos['impact'] * hypos['confidence']) / hypos['efforts']
hypos['ice_rnk'] = hypos['ice'].rank(ascending = False).astype('int')
hypos.sort_values(by ='ice', ascending = False).round(2)
```

Топ3 самых приоритетных гипотез согласно фреймворку ICE :
 1. Запустить акцию, дающую скидку на товар в день рождения
 2. Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей
 3. Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок

#### 2. Примените фреймворк RICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета.

Формула расчета фреймворка RICE:

$$RICE = \frac{Rearch * Impact * Confidence}{Efforts}$$

```python
hypos['rice'] = (hypos['reach'] * hypos['impact'] * hypos['confidence']) / hypos['efforts']
hypos['rice_rnk'] = hypos['rice'].rank(ascending = False, method = 'first').astype('int')
hypos.sort_values(by ='rice', ascending = False).round(2)
```

Топ3 самых приоритетных гипотез согласно фреймворку RICE :
 1. Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок
 2. Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа
 3. Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей

#### 3. Укажите, как изменилась приоритизация гипотез при применении RICE вместо ICE. Объясните, почему так произошло.

```python
hypos_comp = hypos[['hypothesis','ice_rnk','rice_rnk','reach']].copy()
hypos_comp['difference'] = hypos_comp['ice_rnk'] - hypos_comp['rice_rnk']
hypos_comp.sort_values(by = 'difference', ascending = False)
```

Видим, что самая приоритетная гипотеза по фреймворку RICE была только на третьем месте во фреймворке ICE и похожая история с гипотезами имеющий большой охват, ведь именно он так сильно влияет на разницу в результатах. 

В таблице четко видно взаимосвязь между озхватом и разницей в позициях в разных фреймворках. 
Чем больше охват, тем выше разница приоритетов между RICE и ICE

#### 1. Постройте график кумулятивной выручки по группам. Сделайте выводы и предположения.

```python
dates_grouped = orders[['date','group']].drop_duplicates()

orders_agg = dates_grouped\
.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'],orders['group'] == x['group'])]\
.agg({  'date' : 'max'
      , 'group' : 'max'
      , 'transactionId' : 'nunique'
      , 'visitorId' : 'nunique'
      , 'revenue' : 'sum'}), axis=1)\
.sort_values(by=['date','group']) 

orders_agg.columns = ['date','group','tran_un','visit_un','revenue']

plt.figure(figsize=(20, 8))
plt.title('cumulative revenue by groups', fontsize=18)
sns.lineplot(  x="date"
             , y="revenue"
             , hue="group"
             , data=orders_agg)
plt.legend(('A', 'B'),fontsize=20)
plt.grid()
```

#### 2. Постройте график кумулятивного среднего чека по группам. Сделайте выводы и предположения.

```python
visits_agg = dates_grouped\
.apply(lambda x: visits[np.logical_and(visits['date'] <= x['date'], visits['group'] == x['group'])]\
.agg({'date' : 'max'
      , 'group' : 'max'
      , 'visitors' : 'sum'}), axis=1)\
.sort_values(by=['date','group']) 

visits_agg.columns = ['date','group','visit_n']

data_agg = pd.merge(orders_agg,visits_agg, on =['date','group'])
data_agg['avg_bill'] = data_agg['revenue'] / data_agg['visit_n']

avg_bill_mean = data_agg['revenue'].sum() / data_agg['visit_n'].sum()

plt.figure(figsize=(20, 8))
plt.title('avg bill by groups', fontsize=18)
sns.lineplot(  x="date"
             , y="avg_bill"
             , hue="group"
             , data=data_agg)
plt.axhline(y=avg_bill_mean, color='green', linestyle='--', label = 'avg_bill_raw')
plt.legend(loc = 2)
plt.grid()
```

- на график кумулятивного среднего чека по группам снова видно преимущество B над А
- группа А значительно ниже среднего показателя по сырым данным и стабильно хуже группы B

#### 3. Постройте график относительного изменения кумулятивного среднего чека группы B к группе A. Сделайте выводы и предположения.

```python
avg_bill_ratio_mean = data_agg[data_agg['group'] == "B"]['avg_bill'].sum()\
/data_agg[data_agg['group'] == "A"]['avg_bill'].sum()-1


avg_bill_ratio = data_agg.pivot_table(index='date'
                             , columns = 'group'
                             , values = 'avg_bill').reset_index()
avg_bill_ratio['B/A'] = (avg_bill_ratio['B'] / avg_bill_ratio['A'])-1

plt.figure(figsize=(20, 8))
plt.title('avg bill ratio (B/A)', fontsize=18)
sns.lineplot(  x="date"
             , y="B/A"
             , data=avg_bill_ratio)
plt.axhline(y=0, color='black', linestyle='--',label="zero") 
plt.axhline(y=avg_bill_ratio_mean, color='green', linestyle='--',label="avg_bill_ratio_raw")
plt.grid()
plt.legend(loc =2)
```

- группа B стабильно лучше группы A по среднему чеку и различие неравномерно увеличивается со временем
- после скачка разница держится на уровне выше среднего 

#### 4. Постройте график кумулятивной конверсии по группам. Сделайте выводы и предположения.

```python
data_agg['conversion'] = data_agg['tran_un'] / data_agg['visit_n']

avg_conv = data_agg.groupby('date')['conversion'].mean().mean()

plt.figure(figsize=(20, 8))
plt.title('cumulative conversion by groups', fontsize=18)
sns.lineplot(  x="date"
             , y="conversion"
             , hue="group"
             , data=data_agg)
plt.axhline(y=avg_conv, color='green', linestyle='--',label="avg_conversion_raw")
plt.legend(loc = 3)
plt.axis(["2019-08-01", '2019-09-01', 0, 0.04])
plt.grid()
```

#### 10. Посчитайте статистическую значимость различий в конверсии между группами по «сырым» данным. Сделайте выводы и предположения.

Сформулируем гипотезы для нашего теста:

- Нулевая гипотеза: конверсии в группах А и В не отличаются
- Алтернативная гипотеза: конверсии в группах А и В различны

```python
orders_by_users_a = orders.query('group == "A"').groupby('visitorId')['transactionId'].nunique().reset_index()
orders_by_users_a.columns = ['user','orders']

orders_by_users_b = orders.query('group == "B"').groupby('visitorId')['transactionId'].nunique().reset_index()
orders_by_users_b.columns = ['user','orders']

conv_sample_a = pd.concat([orders_by_users_a['orders']
, pd.Series(0,index=np.arange(visits.query('group =="A"')['visitors'].sum()\
                              - len(orders_by_users_a['user'])), name='orders')],axis=0)

conv_sample_b = pd.concat([orders_by_users_b['orders']
, pd.Series(0,index=np.arange(visits.query('group =="B"')['visitors'].sum()\
                              - len(orders_by_users_a['user'])), name='orders')],axis=0)

print("статистическая значимость различий в конверсиях по «сырым» данным:","{0:.5f}".format(stats.mannwhitneyu(conv_sample_a, conv_sample_b)[1]))
print("относительное различие в конверсии между группами по «сырым» данным:","{0:.1%}".format(conv_sample_b.mean()/conv_sample_a.mean()-1)) 
```
