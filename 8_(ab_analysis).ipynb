- Impact ‚Äî –≤–ª–∏—è–Ω–∏–µ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ 10-–±–∞–ª–ª—å–Ω–æ–π —à–∫–∞–ª–µ;
- Confidence ‚Äî —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –≤ –≥–∏–ø–æ—Ç–µ–∑–µ –ø–æ 10-–±–∞–ª–ª—å–Ω–æ–π —à–∫–∞–ª–µ;
- Efforts ‚Äî –∑–∞—Ç—Ä–∞—Ç—ã —Ä–µ—Å—É—Ä—Å–æ–≤ –Ω–∞ –ø—Ä–æ–≤–µ—Ä–∫—É –≥–∏–ø–æ—Ç–µ–∑—ã –ø–æ 10-–±–∞–ª–ª—å–Ω–æ–π —à–∫–∞–ª–µ. –ß–µ–º –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ Efforts, —Ç–µ–º –¥–æ—Ä–æ–∂–µ –ø—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã.


–ò–∑—É—á–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –≥–∏–ø–æ—Ç–µ–∑–∞–º–∏

```python
hypos = pd.read_csv('/datasets/hypothesis.csv')
hypos.columns  = hypos.columns.str.lower()
hypos.info()
```

–î–∞–Ω–Ω—ã–µ –≤ –ø–æ—Ä—è–¥–∫–µ

#### 1. –ü—Ä–∏–º–µ–Ω–∏—Ç–µ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫ ICE –¥–ª—è –ø—Ä–∏–æ—Ä–∏—Ç–∏–∑–∞—Ü–∏–∏ –≥–∏–ø–æ—Ç–µ–∑. –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –∏—Ö –ø–æ —É–±—ã–≤–∞–Ω–∏—é –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞.

–§–æ—Ä–º—É–ª–∞ —Ä–∞—Å—á–µ—Ç–∞ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫–∞ ICE:

$$ICE = \frac{Impact * Confidence}{Efforts}$$

```python
hypos['ice'] = (hypos['impact'] * hypos['confidence']) / hypos['efforts']
hypos['ice_rnk'] = hypos['ice'].rank(ascending = False).astype('int')
hypos.sort_values(by ='ice', ascending = False).round(2)
```

–¢–æ–ø3 —Å–∞–º—ã—Ö –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã—Ö –≥–∏–ø–æ—Ç–µ–∑ —Å–æ–≥–ª–∞—Å–Ω–æ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫—É ICE :
 1. –ó–∞–ø—É—Å—Ç–∏—Ç—å –∞–∫—Ü–∏—é, –¥–∞—é—â—É—é —Å–∫–∏–¥–∫—É –Ω–∞ —Ç–æ–≤–∞—Ä –≤ –¥–µ–Ω—å —Ä–æ–∂–¥–µ–Ω–∏—è
 2. –î–æ–±–∞–≤–∏—Ç—å –¥–≤–∞ –Ω–æ–≤—ã—Ö –∫–∞–Ω–∞–ª–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è —Ç—Ä–∞—Ñ–∏–∫–∞, —á—Ç–æ –ø–æ–∑–≤–æ–ª–∏—Ç –ø—Ä–∏–≤–ª–µ–∫–∞—Ç—å –Ω–∞ 30% –±–æ–ª—å—à–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
 3. –î–æ–±–∞–≤–∏—Ç—å —Ñ–æ—Ä–º—É –ø–æ–¥–ø–∏—Å–∫–∏ –Ω–∞ –≤—Å–µ –æ—Å–Ω–æ–≤–Ω—ã–µ —Å—Ç—Ä–∞–Ω–∏—Ü—ã, —á—Ç–æ–±—ã —Å–æ–±—Ä–∞—Ç—å –±–∞–∑—É –∫–ª–∏–µ–Ω—Ç–æ–≤ –¥–ª—è email-—Ä–∞—Å—Å—ã–ª–æ–∫

#### 2. –ü—Ä–∏–º–µ–Ω–∏—Ç–µ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫ RICE –¥–ª—è –ø—Ä–∏–æ—Ä–∏—Ç–∏–∑–∞—Ü–∏–∏ –≥–∏–ø–æ—Ç–µ–∑. –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –∏—Ö –ø–æ —É–±—ã–≤–∞–Ω–∏—é –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞.

–§–æ—Ä–º—É–ª–∞ —Ä–∞—Å—á–µ—Ç–∞ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫–∞ RICE:

$$RICE = \frac{Rearch * Impact * Confidence}{Efforts}$$

### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞

–°–ø–∞—Å–∏–±–æ –±–æ–ª—å—à–æ–µ,–∑–∞–±—Ä–∞–ª, –≤ —Å–≤–æ–∏ –≤—ã–≤–æ–¥—ã –¥–æ–±–∞–≤–∏–ª, –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –±—É–¥—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–∞–ª—å—à–µ)

```python
hypos['rice'] = (hypos['reach'] * hypos['impact'] * hypos['confidence']) / hypos['efforts']
hypos['rice_rnk'] = hypos['rice'].rank(ascending = False, method = 'first').astype('int')
hypos.sort_values(by ='rice', ascending = False).round(2)
```

–¢–æ–ø3 —Å–∞–º—ã—Ö –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã—Ö –≥–∏–ø–æ—Ç–µ–∑ —Å–æ–≥–ª–∞—Å–Ω–æ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫—É RICE :
 1. –î–æ–±–∞–≤–∏—Ç—å —Ñ–æ—Ä–º—É –ø–æ–¥–ø–∏—Å–∫–∏ –Ω–∞ –≤—Å–µ –æ—Å–Ω–æ–≤–Ω—ã–µ —Å—Ç—Ä–∞–Ω–∏—Ü—ã, —á—Ç–æ–±—ã —Å–æ–±—Ä–∞—Ç—å –±–∞–∑—É –∫–ª–∏–µ–Ω—Ç–æ–≤ –¥–ª—è email-—Ä–∞—Å—Å—ã–ª–æ–∫
 2. –î–æ–±–∞–≤–∏—Ç—å –±–ª–æ–∫–∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–π —Ç–æ–≤–∞—Ä–æ–≤ –Ω–∞ —Å–∞–π—Ç –∏–Ω—Ç–µ—Ä–Ω–µ—Ç –º–∞–≥–∞–∑–∏–Ω–∞, —á—Ç–æ–±—ã –ø–æ–≤—ã—Å–∏—Ç—å –∫–æ–Ω–≤–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –∑–∞–∫–∞–∑–∞
 3. –î–æ–±–∞–≤–∏—Ç—å –¥–≤–∞ –Ω–æ–≤—ã—Ö –∫–∞–Ω–∞–ª–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è —Ç—Ä–∞—Ñ–∏–∫–∞, —á—Ç–æ –ø–æ–∑–≤–æ–ª–∏—Ç –ø—Ä–∏–≤–ª–µ–∫–∞—Ç—å –Ω–∞ 30% –±–æ–ª—å—à–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π

#### 3. –£–∫–∞–∂–∏—Ç–µ, –∫–∞–∫ –∏–∑–º–µ–Ω–∏–ª–∞—Å—å –ø—Ä–∏–æ—Ä–∏—Ç–∏–∑–∞—Ü–∏—è –≥–∏–ø–æ—Ç–µ–∑ –ø—Ä–∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–∏ RICE –≤–º–µ—Å—Ç–æ ICE. –û–±—ä—è—Å–Ω–∏—Ç–µ, –ø–æ—á–µ–º—É —Ç–∞–∫ –ø—Ä–æ–∏–∑–æ—à–ª–æ.

```python
hypos_comp = hypos[['hypothesis','ice_rnk','rice_rnk','reach']].copy()
hypos_comp['difference'] = hypos_comp['ice_rnk'] - hypos_comp['rice_rnk']
hypos_comp.sort_values(by = 'difference', ascending = False)
```

–í–∏–¥–∏–º, —á—Ç–æ —Å–∞–º–∞—è –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ø–æ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫—É RICE –±—ã–ª–∞ —Ç–æ–ª—å–∫–æ –Ω–∞ —Ç—Ä–µ—Ç—å–µ–º –º–µ—Å—Ç–µ –≤–æ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫–µ ICE –∏ –ø–æ—Ö–æ–∂–∞—è –∏—Å—Ç–æ—Ä–∏—è —Å –≥–∏–ø–æ—Ç–µ–∑–∞–º–∏ –∏–º–µ—é—â–∏–π –±–æ–ª—å—à–æ–π –æ—Ö–≤–∞—Ç, –≤–µ–¥—å –∏–º–µ–Ω–Ω–æ –æ–Ω —Ç–∞–∫ —Å–∏–ª—å–Ω–æ –≤–ª–∏—è–µ—Ç –Ω–∞ —Ä–∞–∑–Ω–∏—Ü—É –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞—Ö. 

–í —Ç–∞–±–ª–∏—Ü–µ —á–µ—Ç–∫–æ –≤–∏–¥–Ω–æ –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å –º–µ–∂–¥—É –æ–∑—Ö–≤–∞—Ç–æ–º –∏ —Ä–∞–∑–Ω–∏—Ü–µ–π –≤ –ø–æ–∑–∏—Ü–∏—è—Ö –≤ —Ä–∞–∑–Ω—ã—Ö —Ñ—Ä–µ–π–º–≤–æ—Ä–∫–∞—Ö. 
–ß–µ–º –±–æ–ª—å—à–µ –æ—Ö–≤–∞—Ç, —Ç–µ–º –≤—ã—à–µ —Ä–∞–∑–Ω–∏—Ü–∞ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–æ–≤ –º–µ–∂–¥—É RICE –∏ ICE

#### 4. –í—ã–≤–æ–¥

```python
dates_grouped = orders[['date','group']].drop_duplicates()

orders_agg = dates_grouped\
.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'],orders['group'] == x['group'])]\
.agg({  'date' : 'max'
      , 'group' : 'max'
      , 'transactionId' : 'nunique'
      , 'visitorId' : 'nunique'
      , 'revenue' : 'sum'}), axis=1)\
.sort_values(by=['date','group']) 

orders_agg.columns = ['date','group','tran_un','visit_un','revenue']

plt.figure(figsize=(20, 8))
plt.title('cumulative revenue by groups', fontsize=18)
sns.lineplot(  x="date"
             , y="revenue"
             , hue="group"
             , data=orders_agg)
plt.legend(('A', 'B'),fontsize=20)
plt.grid()
```
